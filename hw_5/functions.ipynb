# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

#---------Решение---------
debug = True #Для дебага

directories={} 
directories[1]=[]
directories[2]=[]
directories[3]=[]

for element in documents:
    if element['type'] == 'passport' or element['type'] == 'invoice':
        directories[1].append(element['number'])
        if debug:print(1, directories[1])
    elif element['type'] == 'insurance':
        directories[2].append(element['number'])
        if debug:print(2, directories[2])
    else: print('Господин, документы проверены. Карта Грааля не найдена. Поищите обелиск с картой.')

if debug:print()
    
for key,data in directories.items():
    print("{0}: {1}".format(key,data))   

#-------------- Задание 1. Пункт 1. Функция "Найти имя" --------------    

def find_name():
    """
        Ищет имя в базе 'documents' по номеру документа
    """
    user_number = input('Введите номер документа:')
    flag = False
    
    for element in documents:
        if element['number'] == user_number:
            print('Владелец документа: ', element['name'])
            flag = True
    if flag == False: print('Документ не найден в базе')
        
    
    return None    

#find_name()

#-------------- Задание 1. Пункт 2. Функция "Найти полку" --------------  

def find_shelf(print_results=True, number=None):
    """
        Ищет полку в базе 'directories' по номеру документа
        print_results=True - выводит на экран служебную информацию
        number=None - начальное число
    """
    if number == None:
        user_number = input('Введите номер документа:')
    else: user_number = number
        
    flag = False
    
    for key, data in directories.items():
        if list(filter(lambda x: True if x == user_number else False , data)):
            if print_results: print('Документ хранится на полке: ', key)
            flag = True
            return key
    if flag == False: 
        if print_results: print('Документ не найден в базе')
    return None

#find_shelf()
#find_shelf(print_results=True, number='11-2')

#-------------- Задание 1. Пункт 3. Функция "Вспомнить все: полную информацию по всем документам" -------------- 

def total_recall(prints='Yes'):
    """
        Выдает полную информацию по всем документам в базе 'documents', а также полку в базе 'directories' 
    """
    for element in documents:        
        passport, number ,name = list(element.values()) 
        print(f'№: {number}, тип: {passport}, владелец: {name}, полка хранения: {find_shelf(print_results=False, number=number)}')

    return None

#total_recall()

#-------------- Задание 1. Пункт 4. Функция "Добавить полку" -------------- 
#Такая полка уже существует. Текущий перечень полок: 1, 2, 3.

def add_shelf():
    """
        Добавит полку в базу 'directories'. Выведет результат
        Проведет проверку если Такая полка уже существует. 
    """
    new_shelf = Do_user_input()
    if new_shelf not in directories:
        directories[new_shelf]=[]  # Создаем полку
        print('Полка добавлена.')
    else: 
        print('Такая полка уже существует.', end=" ") 
    print('Текущий перечень полок:', end=" ")
    print_result()

def print_result():
    """
        Составлет перечень ключей базы 'directories' формата"1, 2, 3, 135."
    """
    k=0
    for i in directories.keys():
        k +=1
        if k == len(directories.keys()):
            print(i, end=".")
        else:print(i, end=", ")  

def Do_user_input(string="Введите номер полки: "):
    """
        Запрашивает ввод пользователя (в цикле while) и Проверяет ввел ли пользователь число, а не что то другое. 
        Возвращает ввод пользователя - если число. 
    """
    while True:
        value = input(string)
        if value.isalpha() or int(value) <= 0: continue
        else: return int(value)
        

#Do_user_input()  
#print_result()
#add_shelf()


#-------------- Задание 1. Пункт 5. Функция "Удалить полку" -------------- 

def you_were_my_brother_Anakin():
    """
        Удалит полку из базы 'directories', если там ничего нет. Выведет результат
    """
    user_input = Do_user_input()
    if user_input not in directories: 
        print("Такой полки не существует. ", end="")
        
    elif not directories[user_input]: 
        del(directories[user_input])
        print("Полка удалена. ", end="")
    
    elif directories[user_input] is not None:    
        print("На полке есть документы, удалите их перед удалением полки. ", end="")
        
    print(" Текущий перечень полок: ", end="")
    print_result()
    print()    
    return None 

#you_were_my_brother_Anakin()


#-------------- Задание 2. Пункт 1. Функция "добавить новый документ" -------------- 

def add_new_id():
    """
        добавить новый документ в базы 'directories' и 'documents'
    """ 
    id_number = input("Введите номер документа:") 
    id_type = input("Введите тип документа:")
    id_newUser = input("Введите владельца документа:")
    id_shelf = int(Do_user_input("Введите полку для хранения:"))
    
    if id_shelf not in directories: print("Такой полки не существует. Добавьте полку командой ads.")
    else: 
        directories[id_shelf].append(id_number)
        A={}
        A['type'] = id_type
        A['number'] = id_number
        A['name'] = id_newUser
        documents.append(A)
        print("Документ добавлен.",end="")
    print("Текущий список документов:")
    total_recall()
    
#add_new_id()


#-------------- Задание 2. Пункт 2. Функция "удалить документ" -------------- 
                
def erase_by_id_number():
    """
        удалить документ по номеру из базы 'directories' и 'documents'
    """ 
    id_number = input("Введите номер документа:")
    shelf = find_shelf(print_results=False, number=id_number)
    if shelf == None: 
        print("Документ не найден в базе.")
    else:
        directories[shelf].remove(id_number) #Удален из полки
        i=0
        for element in documents:
            if element['number'] == id_number:
                del(documents[i])           #Удален из documents
                break
            i +=1
        print("Документ удален.")
        
    print("Текущий список документов:")
    total_recall() 
    
#erase_by_id_number()   


#-------------- Задание 2. Пункт 3. Функция "Переместить документ" -------------- 
    
def move_by_id_and_shelf():
    """
        Переместить документ по номеру  и номеру полки 
    """ 
    id_number = input("Введите номер документа:")    
    new_shelf = int(Do_user_input())

    shelf = find_shelf(print_results=False, number=id_number)
    
    if new_shelf not in directories: 
        print("Такой полки не существует. Текущий перечень полок: ", end="")
        print_result()
        return None
    elif shelf == None: 
        print("Документ не найден в базе.")
    else: 
        directories[shelf].remove(id_number)     #Удален из полки
        directories[new_shelf].append(id_number) #Добавлен на новую полку 
        print("Документ перемещен.")
        
    print("Текущий список документов:")
    total_recall()
    return None

#move_by_id_and_shelf()


#-------------- Финальная функция main() -------------- 

def main():
    while True:
        user_input = input('Введите команду:')
        if user_input == 'p': find_name()
        if user_input == 's': find_shelf()
        if user_input == '/': total_recall()               #Выведет все полки
        if user_input == 'ads': add_shelf()
        if user_input == 'ds': you_were_my_brother_Anakin()    #Удалит полку
        if user_input == 'ad': add_new_id()                #Добавит Нового Пользователя
        if user_input == 'd': erase_by_id_number()         #Удалит по номеу документа
        if user_input == 'm': move_by_id_and_shelf()       #Переместит по номеу документа
        elif user_input == 'q': return None
main()
        
````
