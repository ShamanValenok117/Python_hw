## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера.  
Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
debug = False        #Дебаг

ratings = pd.read_csv("dataset/ratings.csv")


a = 0                #Служебная. Подсчет макс.
movie_id = ''        #Служебная. Номер макс. фильма
if debug == False: print('Выполняю подсчет ',end="")

for element in ratings[ratings.rating == 5].movieId.unique():
    b = ratings[ (ratings.rating == 5) & (ratings.movieId == element) ]['userId'].count()
    if debug:
        print(f'Фильм id = {element}, получил {b} рейтингов 5')
    if b > a: 
        a = b
        movie_id = element
    if debug == False: print('.',end="")
print()
print(f'Фильм с movieId = {movie_id} имеет больше всего рейтингов 5.00')
    
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
electric = pd.read_csv("csv_files/power.csv")

#country = Lithuania Estonia Latvia 
filtered_countries = electric[ (electric['country'] == 'Lithuania') | (electric['country'] == 'Estonia') | (electric['country'] == 'Latvia') ]

# category = 4  12  21
filtered_countries_category = filtered_countries[ (filtered_countries['category'] == 4) | (filtered_countries['category'] == 12) |(filtered_countries['category'] == 21) ]

#c 2005 по 2010 года
filtered_countries_category_timed = filtered_countries_category[ (filtered_countries_category['year'] >= 2005 ) & (filtered_countries_category['year'] <= 2010 )]

# Не учитывайте в расчетах отрицательные значения quantity.
filtered_countries_category_timed_quantity = filtered_countries_category_timed[ filtered_countries_category_timed['quantity'] > 0 ]

filtered_countries_category_timed_quantity['quantity'].count()
```

```python
# ---- Решение 2. упрощенное ----

electric = pd.read_csv("csv_files/power.csv")

#country = Lithuania Estonia Latvia 
filtered_countries =electric[ electric['country'].isin(['Lithuania','Estonia','Latvia'])]

# category = 4  12  21
filtered_countries_category = filtered_countries[ filtered_countries['category'].isin([4,12,21]) ]  

#c 2005 по 2010 года
filtered_countries_category_timed = filtered_countries_category[ filtered_countries_category['year'].isin([2005,2006,2007,2008,2009,2010])]

# Не учитывайте в расчетах отрицательные значения quantity.
filtered_countries_category_timed_quantity = filtered_countries_category_timed[ filtered_countries_category_timed['quantity'] > 0 ]

filtered_countries_category_timed_quantity['quantity'].count()
```
