## Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:
- среднюю букву, если число букв в слове нечетное;
- две средних буквы, если число букв четное.

Примеры работы программы:  
1. `word = 'test'`

Результат:  
`es`
2.     
`word = 'testing'`

Результат:  
`t`

````python
word = 'testing'
string1=str(word)

if len(string1)%2==0:      #если четное
    p=int(len(string1)/2)
    print(string1[p-1:p+1])
else:                      #если не четное
    p=int(len(string1)/2)
    print(string1[p:p+1])
````

## Задание 2
Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

Примеры работы программы:  
1.   

```
Введите число:  
1

Введите число:  
4

Введите число:  
6

Введите число:  
0
```

Результат:  
`11`

2.      

```
Введите число:  
0
```  

Результат:  
`0`

````python
flag = False
user_input = []

while flag==False:
    user_input.append(int(input('Введите число: ')))
    if user_input[-1]==0:flag=True         #Проверка последнего ввода значения "0"

print('Результат:',sum(user_input) )
````

````python
# Улучшенная версия: с проверкой на число
flag = False
user_input = []

while flag==False:
    n=input('Введите число: ')
    if n.isalpha(): continue #Проверка на то, что это число

    if int(n)==0:flag=True
    user_input.append(float(n))

print('Результат:',sum(user_input) )
````

## Задание 3

Мы делаем [MVP](https://ru.wikipedia.org/wiki/%D0%9C%D0%B8%D0%BD%D0%B8%D0%BC%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE_%D0%B6%D0%B8%D0%B7%D0%BD%D0%B5%D1%81%D0%BF%D0%BE%D1%81%D0%BE%D0%B1%D0%BD%D1%8B%D0%B9_%D0%BF%D1%80%D0%BE%D0%B4%D1%83%D0%BA%D1%82) dating-сервиса, и у нас есть список парней и девушек.  
Выдвигаем гипотезу: лучшие рекомендации мы получим, если просто отсортируем имена по алфавиту и познакомим людей с одинаковыми индексами после сортировки!
Но мы не будем никого знакомить, если кто-то может остаться без пары:

Примеры работы программы:  
1.  

```
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']

```

2.     

```

```

Результат:  
`Внимание, кто-то может остаться без пары!`

````python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

print('Результат:')

if len(boys)>len(girls) or len(boys)<len(girls): print('Внимание, кто-то может остаться без пары! ') 
else:
    print('Идеальные пары: ')
    for i in range(len(boys)):
        print (sorted(boys)[i]+ ' и '+ sorted(girls)[i])
````

## Задание 4

У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам (структура данных в примере).
Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.

Пример работы программы:  
```
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
```

Результат:  
```
Средняя температура в странах:
Thailand  -  23.9 С
Germany  -  13.8 С
Russia  -  3.7 С
Poland  -  12.0 С
```

## (Фаренгейт — 32) : 1,8 = Цельсий 
Пример: (50°F - 32) : 1,8 = 10°C 

````python
print('Средняя температура в странах:')
for i in range(len(countries_temperature)):
    F = sum(countries_temperature[i][1])/len(countries_temperature[i][1])
    C = round((F - 32) / 1.8,2)
    print(countries_temperature[i][0], ' - ', C, '°C')
````

## Задание 5 (необязательное)

Дан поток логов по количеству просмотренных страниц для каждого пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. 
Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

Примеры работы программы:  
1.  

```
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]
```

Результат:  
`Среднее количество просмотров на уникального пользователя: 23.25`
2.     

```
stream = [
    '2018-01-01,user100,150',
    '2018-01-07,user99,205',
    '2018-03-29,user1001,81'
]
```

Результат:  
`Среднее количество просмотров на уникального пользователя: 145.33`

````python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]

A=[] # разбиваем строку в список
for i in range(len(stream)):
    A.append(stream[i].split(','))

    
Users = set([element[1] for element in A]) # получаем уникальные имена пользователей - перевели второй столбец во множество
    
Sum1 = sum([int(element[2]) for element in A]) #сумма просмотров всего

User_average=round(Sum1/len(Users),2) # Cреднее количество просмотров на уникального пользователя
 

print('Среднее количество просмотров на уникального пользователя: ',User_average)
````

## Задание 6 (необязательное)

Необходимо у пользователя запрашивать набор чисел разделенных пробелом. В результате должны выводиться числа, которые повторяются в вводе более одного раза.

Примеры работы программы:  
1.  

```
Введите числа:
4 8 0 3 4 2 0 3
```

Результат:  
`0 3 4`
2.     

```
Введите числа:
1 1 2 2 3 3
```

Результат:  
`1 2 3`
3.     

```
Введите числа:
1 1 1 1 1 2 2 2
```

Результат:  
`1 2`

````python
user_input=input('Введите числа:').split(' ')

B = set(user_input) #делаем множество уникалььных значений

for element in B:
        E = user_input.count(element)   #Считаем количество повторений и выписываем все, где повторений больше 1
        if E > 1: print(element)
````
